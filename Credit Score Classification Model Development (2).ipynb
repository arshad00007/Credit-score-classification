```python
num_col1.head()
```

```python
# Identify the outliers and remove 

for i in num_col1:
    Q1=train_df1[i].quantile(0.25) # 25th quantile
    Q3=train_df1[i].quantile(0.75) # 75th quantile
    IQR = Q3-Q1
    Lower_Whisker = Q1 - 1.5*IQR 
    Upper_Whisker = Q3 + 1.5*IQR
    train_df1[i] = np.clip(train_df1[i], Lower_Whisker, Upper_Whisker)
```

```python
# PLot the numerical columns
plt.figure(figsize=(20,30))
for i, variable in enumerate(num_col):
                     plt.subplot(5,4,i+1)
                     plt.boxplot(train_df1[variable],whis=1.5)
                     plt.tight_layout()
                     plt.title(variable)
```

This code will generate a grid of box plots, each representing a different numerical column in our DataFrame (train_df1). It's a useful way to visualize the distribution and potential outliers in our data.
